```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('DataAnalyst.csv',encoding='gb2312')
```

```python
len(df.positionId.unique())
```

```python
df_duplicates = df.drop_duplicates(subset='positionId',keep='first')
```

```python
df_duplicates.head()
```

```python
def cut_word(word):
    position = word.find('-')
    if position != -1:
        bottomSalary = word[:position-1]
    else:
        bottomSalary = word[:word.upper().find('K')]
    return bottomSalary
```

```python
df_duplicates['bottomSalary'] = df_duplicates.salary.apply(cut_word)
```

```python
df_duplicates.head()
```

```python
def top_word(word):
    position = word.find('-')
    if position != -1:
        topSalary = word[position+1:-1]
    else:
        topSalary = word[:word.upper().find('K')]
    return topSalary
```

```python
def cut_word(word,method):
    position = word.find('-')
    if position != -1:
        bottomSalary = word[:position-1]
        topSalary = word[position+1:-1]
    else:
        bottomSalary = word[:word.upper().find('K')]
        topSalary = bottomSalary
    if method == 'bottom':
        return bottomSalary
    else:
        return topSalary
```

```python
df_duplicates['topSalary']=df_duplicates.salary.apply(cut_word,method='top')
```

```python
df_duplicates.head()
```

```python
df_duplicates['bottomSalary'] = df_duplicates.bottomSalary.astype('int')
df_duplicates['topSalary'] = df_duplicates.topSalary.astype('int')
```

```python
df_duplicates['avgSalary'] = df_duplicates.apply(lambda x:(x.topSalary+x.bottomSalary)/2,axis=1)
```
