```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取数据
train=pd.read_csv("cmc_train.csv")
test=pd.read_csv("test.csv")
```

```python
train.info()
```

```python
# 2 与类别的关系   妻子受教育程度1 2 3 4 跟使用避孕方法
#  2   4中
#  1.2  3种
counts = np.zeros((4,3))
for data in range(len(train)):
    counts[train.values[data][2]-1][train.values[data][-1]-1]+=1

counts=counts.T
width = 0.2
x = [i-0.2 for i in range(1,5)]
plt.bar(x,counts[0],width=width)

x = [i-0.0 for i in range(1,5)]
plt.bar(x,counts[1],width=width)

x = [i+0.2 for i in range(1,5)]
plt.bar(x,counts[2],width=width)
```

```python
# 3 与 1.2的关系  丈夫教育 与  使用避孕方法

counts = np.zeros((4,3))
for data in range(len(train)):
    counts[train.values[data][3]-1][train.values[data][-1]-1]+=1

counts=counts.T
width = 0.2
x = [i-0.2 for i in range(1,5)]
plt.bar(x,counts[0],width=width)

x = [i-0.0 for i in range(1,5)]
plt.bar(x,counts[1],width=width)

x = [i+0.2 for i in range(1,5)]
plt.bar(x,counts[2],width=width)
```
