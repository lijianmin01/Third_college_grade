```python
import matplotlib.pyplot as plt 
import numpy as np

plt.plot([1,2,3,4])
```

```python

x = np.linspace(0,6,1000)

y_l = np.arange(-1.0,0.0,0.2)

f1= np.sin(x)
f2 =np.cos(x)

plt.figure()

plt.plot(x,f1,'r-.',linewidth=3)
plt.plot(x,f2,'g-',linewidth=3)
plt.ylim(-1.2,0.0)
plt.grid(True)

label_name = ['sin(x)', 'cos(x)']
plt.legend(label_name, loc=1)

plt.show()
```
