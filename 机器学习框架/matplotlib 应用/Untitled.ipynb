```python
import matplotlib.pyplot as plt 
import numpy as np

plt.plot([1,2,3,4])
```

```python

x = np.linspace(0,6,1000)

y_l = np.arange(-1.0,0.0,0.2)

f1= np.sin(x)
f2 =np.cos(x)

plt.figure()

plt.plot(x,f1,'r-.',linewidth=3)
plt.plot(x,f2,'g-',linewidth=3)
plt.ylim(-1.2,0.0)
plt.grid(True)

label_name = ['sin(x)', 'cos(x)']
plt.legend(label_name, loc=1)

plt.show()
```

```python
x = np.arange(1,10,0.2)
y = np.sin(x)

plt.xlim(0,6)
plt.ylim(-1,0)

plt.xlabel('x')
plt.ylabel('sin(x),cos(x)')
plt.title('plt sin(x)\cos(x)')

plt.grid(True)


plt.plot(x,y,'r-.',linewidth=3)
#plt.plot(x,f2,'g-',linewidth=3)
```

```python
x = np.random.normal(0,1,100)
y = np.random.normal(0,1,100)
x1 = np.random.normal(0,2,100)
y1 = np.random.normal(0,2,100)

plt.figure(figsize=(12,8))
plt.scatter(x,y,marker='.',color='blue',linewidths=3)
plt.scatter(x1,y1,marker='+',color='r',s=100)
plt.grid(True)

label = ['scale=1','scale=2']
plt.legend(label,loc=1)

plt.show()
```

```python
x = [5,8,10,6,9,11]
x=[str(i) for i in x]
y1=[12,15,6]
y2=[6,14,7]
plt.figure()
plt.bar(x[:3],y1)
plt.bar(x[3:],y2,color='g')
plt.title('Bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
data ={
    'a':np.arange(50),
    'd':np.random.randn(50)
}

plt.figure()
plt.scatter('a','d',data=data)
plt.show()
```

```python
data ={
    'a':np.arange(50),
    'd':np.random.randn(50)*100,
    'b':10*np.random.randn(50)
}

plt.figure()
plt.scatter('a','d',data=data)
plt.show()
```
